```
from pyspark.sql import SparkSession
from pyspark.sql.types import *   
    
schema = StructType([ 
    StructField("MARKET",StringType(),True), 
    StructField("SERIES",StringType(),True), 
    StructField("SYMBOL",StringType(),True), 
    StructField("SECURITY", StringType(), True), 
    StructField("PREV_CL_PR", FloatType(), True), 
    StructField("OPEN_PRICE", FloatType(), True),
    StructField("LOW_PRICE", FloatType(), True),
    StructField("CLOSE_PRICE", FloatType(), True),
    StructField("NET_TRDVAL", FloatType(), True),
    StructField("CORP_IND", FloatType(), True),
    StructField("TRADES", FloatType(), True),
    StructField("HI_52_WK", FloatType(), True),
    StructField("LO_52_WK", FloatType(), True)    
  ])
    
```

```
df=spark.read.csv("/FileStore/tables/sharemarket.csv",schema=schema)
```

```
df.head(5)
```

```
df.createOrReplaceTempView('market')
```

```
%sql select * from market limit 10;
```

```
%sql select SERIES,count(*) from market group by SERIES
```

```
%sql select (SERIES) from market
```

```
%sql SELECT series, ROUND(SUM(PREV_CL_PR),2) AS Sum_Prev_Cl, ROUND(SUM(OPEN_PRICE),2) AS Sum_open_price,ROUND(SUM(LOW_PRICE),2) AS Sum_low_price, ROUND(SUM(CLOSE_PRICE),2) AS Sum_close_price  FROM market GROUP BY series
```

```
%sql select security,series from market where NET_TRDVAL = (select max(NET_TRDVAL) from market)
```

```
%sql SELECT SERIES FROM market WHERE (high_price+low_price+open_price)> net_trdval 
```

```
%sql select series from market where net_trdval = (select max(net_trdval)from market)
```

```
%sql SELECT MAX(open_price) AS MAX_OPEN_PRICE, MIN(open_price) AS MIN_OPEN_PRICE FROM market
```

```
%sql SELECT DISTINCT series FROM market where trades > 80
```

```
%sql SELECT SERIES,round(SUM(NET_TRDVAL)-SUM(TRADES),2) AS DIFF FROM market GROUP BY SERIES ORDER BY DIFF DESC
```
